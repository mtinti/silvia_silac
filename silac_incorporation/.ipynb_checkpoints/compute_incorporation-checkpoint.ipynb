```python
import os
import pandas as pd
import numpy as np
from scipy import stats
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
plt.style.use('ggplot')
current_palette = sns.color_palette()
sns.palplot(current_palette)
```

```python
indata = 'peptides.txt.gz'
to_use =['K Count', 'R Count']+['Ratio H/L']

pep = pd.read_table(indata, low_memory=False, usecols=to_use)
np.log2(pep['Ratio H/L']).plot(kind='hist')
```

```python
def plot_incorporation(pep, ratio='M/L', exp='A', ax=''):
    temp = pep[(pep['K Count']==0) & (pep['R Count']>=1)]
    #print (temp.head())
    temp = 1/temp['Ratio '+ratio]
    temp = 1-temp
    temp = temp[(temp>0.6)&(temp<1)]
    temp.plot(kind='hist', ax=ax, bins=50, label='R',color=current_palette[2])
    #ax[0].set_title('R')
    temp = pep[(pep['R Count']==0) & (pep['K Count']>=1)]
    temp = 1/temp['Ratio '+ratio]
    temp = 1-temp
    temp = temp[(temp>0.6)&(temp<1)]
    temp.plot(kind='hist',ax=ax, bins=50, label='K',color=current_palette[3])
    #ax[1].set_title('K')
    
    
def plot_incorporation2(pep, ratio='H/L',ax=ax):
    merge = []
    temp = pep[(pep['K Count']==0) & (pep['R Count']>=1)]
    #print(temp.head())
    temp['Ratio '+ratio] = 1/temp['Ratio '+ratio]
    temp['Ratio '+ratio] = 1-temp['Ratio '+ratio]
    temp['aminoacid']='R'
    merge.append(temp)
    #print (temp.head())
    temp = pep[(pep['R Count']==0) & (pep['K Count']>=1)]
    temp['Ratio '+ratio] = 1/temp['Ratio '+ratio]
    temp['Ratio '+ratio] = 1-temp['Ratio '+ratio]
    temp['aminoacid']='K'
    merge.append(temp)
    merge = pd.concat(merge)
    sns.boxplot(data=merge,x='aminoacid',y='Ratio '+ratio,showfliers=False,ax=ax)   
```

```python
fig,ax=plt.subplots(ncols=1, nrows=1,figsize=(8,6))    
#plot_incorporation(pep, ratio='M/L', exp='A', ax=ax[0]) 
plot_incorporation(pep, ratio='H/L', exp='A', ax=ax)
ax.set_title('Heavy')
ax.set_xlabel('Incorporation Rate', fontsize=12)
plt.legend()
plt.tight_layout()
fig.savefig(os.path.join('Figure_1.svg'))
fig.savefig(os.path.join('Figure_1.png'))
plt.show()
```

```python
fig,ax=plt.subplots(ncols=1, nrows=1,figsize=(8,6)) 
plot_incorporation2(pep, ax=ax)
ax.set_title('Heavy')
ax.set_xlabel('Incorporation Rate', fontsize=12)
plt.tight_layout()
fig.savefig(os.path.join('Figure_2.svg'))
fig.savefig(os.path.join('Figure_2.png'))
plt.show()
```
